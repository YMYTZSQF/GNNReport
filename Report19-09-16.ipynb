To compare with GraphRNN, I have generated 100 samples of grids of different sizes (10\*10 to 19\*19). 
I have used pytorch-biggraph to generate embeddings for each graph and this embedding on node-level. The reason to use node-level embedding is that if we combine all node-level embeddings into one graph-level embedding and train with this vector, it's hard to re-generate graphs based on the output vector.

The input of GraphRNN graph-level rnn is an adjusted adjacent matrix: (1) The order of node is the BFS order; (2) Each node only considers 40 nodes in front of it. 

But this adjusted adjacent matrix can be viewed as a embedding of the graph, each row is a binary vector which represents the embedding of that node. We can use embeddings generated by pytorch-biggraph to replace these embeddings.
