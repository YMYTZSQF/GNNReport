To compare with GraphRNN, I have generated 100 samples of grids of different sizes (10\*10 to 19\*19). 
I have used pytorch-biggraph to generate embeddings for each graph and these embeddings are on node-level. The reason to use node-level embedding is that if we combine all node-level embeddings into one graph-level embedding and train with this vector, it's hard to re-generate graphs based on the output vector.

I have thought about how to integrate the embeddings generated by pytorch-biggraph into the GraphRNN. Currently, I think there are two options:
![image.png](img/Report19-09-16_1.jpg)
The input of GraphRNN graph-level rnn is an adjusted adjacent matrix: (1) The order of node is the BFS order; (2) Each node only considers 40 nodes in front of it. 

But this adjusted adjacent matrix can be viewed as a embedding of the graph, each row is a binary vector which represents the embedding of that node. We can use embeddings generated by pytorch-biggraph to replace these embeddings.

![image.png](img/Report19-09-16_2.jpg)
The second option is to replace the output matrix of the graph-level rnn. The output matrix is n\*16. We can view the function of the graph-level rnn is to get a more abstract embedding from the input adjacent matrix. Therefore, we can replace this embedding with the pytorch-biggraph generated embeddings. 

The advantage of the first option is that the one more level of rnn may capture more useful feature and generate more powerful embeddings while the advantage of the second one is that since we skip the first rnn, there will be large savings in GPU memory costs.
