I have found out that the GraphRNN's original generated graphs are actually connected. 

Previously, we have observed that the generated graphs are much smaller than the maximum size we set. In their codes, they picked the largest connected component. When I dived into the codes and I found the originally the whole generated graph are in most cases are actually connected! The reason that we observed much smaller generated graphs is because that they do a strange operation that for each graph, they check nodes in increasing order of their index, when it finds a node k that has no edge with node 1 to node k-1,  then they substract the subgraph with node 1 to node k - 1, and then return the maximum connected component as the representative of this graph. 

I do not understanding the meaning of doing that because even node  k has no edge to any node in front of it (node 1 to node k - 1), it still could be connected to them by other nodes, for example node 1 has edge to node k + 2 and node k also has edge to node k + 2. So the connected component can be much larger than the subgraph with node 1 to node k-1. The second step is also confusing. Since for each node from 1 to k-1, it must has at least one edge to node in front of it (except node 1), the subgraph with node 1 to node k - 1 must be a connected component. There is no need to return the maximum connected component of that subgraph.

By removing these confusing operations and returning the original generated graph. The generated graph is connected. I only saw few generated graphs with one isolated node and the other part are connected(3600 nodes).

Since we can generate connected large scale graphs, the problem becames what is the quality of these generated large graphs.
